***Creating a class with outside variables and returning a single card with it's rank and suit.***

```python
import random
```

```python
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
class Card:
    
    def __init__(self,suit,rank):
        
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        
        return self.rank + ' of ' + self.suit
    
```

```python
card1 = Card(suits[0],ranks[12])
```

***This class will make a deck of all cards and it will shuffle the cards randomly.***

```python
class Deck:
    
    def __init__(self):
        
        self.allcards = []
        for suit in suits:
            for rank in ranks:
                self.allcards.append(Card(suit,rank))
                
    def shuffle(self):
        random.shuffle(self.allcards)
        
    def deal_one(self):
        return self.allcards.pop()
```

***Creating a Player Class to hold, remove and add number of cards with any possibilities that can occur in the game.***

```python
class Player:
    
    def __init__(self,name):
        
        self.name = name
        self.allcards = []
        
    def remove_one(self):
        
        # Assuming the very left as top of the cards ie. index 0 and bottom of the cards as index -1
        return self.allcards.pop(0)
        
    def add_cards(self,new_cards):
        
        #for the new_cards as list use extend (not to create nested lists)
        if type(new_cards) == type([]):
            self.allcards.extend(new_cards)
        else:
            self.allcards.append(new_cards)
    
    def __str__(self):
        
        return f"Player {self.name} has {len(self.allcards)} cards."
```

    game_on = True

        either one is 0, game is over.
    war = True

        player 1 card > player 2 card,
        all card goes to player 1,
        break from war loop
        case 2:
        player 2 card > player 1 card,
        all card goes to player 2,
        break from war loop
        case 3:
        player 1 card = player 2 card,
        *WAR*
        if player 1 has < 5 cards then player 2 wins,
        if player 2 has < 5 cards then player 1 wins,
        else draw 5 cards and check!
        break from war loop***

```python
game_on = True
counter = 0

while game_on:
    
    counter += 1
    print(f"Round {counter}")
    
    if len(player_1.allcards) == 0:
        print("Player 1 is out of cards! Player 2 wins!")
        game_on = False
        break
        
    if len(player_2.allcards) == 0:
        print("Player 2 is out of cards! Player 1 wins!")
        game_on = False
        break
        
    player_1_cards = []
    player_1_cards.append(player_1.remove_one())
    
    
    player_2_cards = []
    player_2_cards.append(player_2.remove_one())
    
    war = True
    
    while war:
        
        if (player_1_cards[-1].value) > (player_2_cards[-1].value):
            
            player_1.add_cards(player_1_cards)
            player_1.add_cards(player_2_cards)
            war = False
            
        elif (player_1_cards[-1].value) < (player_2_cards[-1].value):
            
            player_2.add_cards(player_1_cards)
            player_2.add_cards(player_2_cards)
            war = False
            
        else:
            
            print("WAR")
            
            if len(player_1.allcards) < 5:
                print("Player 1 is out of cards! Player 2 wins!")
                game_on = False
                break
            
            elif len(player_2.allcards) < 5:
                print("Player 2 is out of cards! Player 1 wins!")
                game_on = False
                break
                
            else:
                
                for num in range(5):
                    
                    player_1_cards.append(player_1.remove_one())
                    player_2_cards.append(player_2.remove_one())
                
```
